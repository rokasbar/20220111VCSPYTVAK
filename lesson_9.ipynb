# Pandas, Matplotlib & Seaborn - duomenu vizualizavimas, grafiku braizymo paketai

```python
# Susiinstaliuojam pip install matplotlib
# ir importuotjam Matplotlib

import matplotlib.pyplot as plt
```

```python
x = [0,100]
y = [0, 200]
```

```python
fig = plt.figure()
#sukure figura ja spausdinam ir pridedam visokiu parametru
plt.plot(x,y)
plt.title('First plot')
plt.xlabel('x axis')
plt.ylabel('y axis')
```

```python
fig= plt.figure()
ax = fig.add_axes([ 0, 0, 1, 1])  #grafikui uzdejo koordinates
ax.plot(x,y)
ax.set_title('First plot')
ax.set_xlabel('x axis')
ax.set_ylabel('y axis')

ax_2 = fig.add_axes( [0.2, 0.3, 0.2, 0.3])
# rect [ x_start, y_start, width, height]
#pirmas skaicius - pozicija X asimi
#antras skaicius - pozicija Y asimi
#trecias skaicius - Plotis
#ketvirtas skaicius - aukstis

ax_2.plot(x,y)

ax_2.set_xlim(50, 60)
ax_2.set_ylim(2000, 5000)
plt.show()

```

# NUMPY biblioteka - tinka su dideliu kiekiu duomenu

```python
import numpy as np
```

```python
x = np.arange(0,100)
print(x)
print(type(x))
y = x * 2
print(y)
```

#griztam atgal i matplotlib

#Jei norim keliu grafiku salia

```python
fig, axes = plt.subplots(1, 2)
axes [0].plot(x, x, ls = '--', lw = 5)
axes [0].plot(x, x, ls = '--', lw = 5)
axes [1].plot(x, y, c = 'r')
```

```python
fig, axes = plt.subplots(3, 3)
axes[1, 2].plot (x, y) #eile indeksas ir stulpelio indeksas
plt.tight_layout() #sustumbodo kad grafikas neliptu ant kitu

for row in axes:
    row[1].plot(x,x)
    for column_ax in row:
        column_ax.plot(x,y)
```

# SEABORN

```python
import seaborn as sns

#atsisiunciam DATA apie ARBATPINIGIUS
tips = sns.load_dataset('tips')
tips
```

```python
sns.countplot(data = tips, x = 'size')
#matosi iskart kad didzioji dalis klientu ateina dviese
#nurodom duombaze, kuri stulpeli imam
```

```python
# Kitas grafiko tipas, dar kas eina i Y asi
sns.jointplot(data = tips, x = 'size', y = 'total_bill')
```

```python
sns.jointplot(data = tips, x = 'size', y = 'total_bill', kind = 'reg')
```

```python
#Atskirimas pagal LYTI
sns.jointplot(data = tips, x = 'size', y = 'total_bill', hue = 'sex')
```

```python
sns.boxplot(data = tips, x = 'smoker', y = 'tip')
```

```python
sns.scatterplot(data = tips, x = 'total_bill', y = 'tip', hue = 'size', size = 'time')
```

```python
#Padidina tapacia figura
plt.figure (figsize=(15,10) )
sns.scatterplot(data = tips, x = 'total_bill', y = 'tip', hue = 'size', size = 'time')
```
