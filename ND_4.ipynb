```python
import pandas as pd
```

```python
df = pd.read_csv('LT.csv')
df
```

```python
df.rename(columns = {'tourniment': 'tournament'}, inplace = True)
```

```python
#Sukurti naują stulpelį, kuriame būtų paskaičiuota įvarčiu suma
df['score_total'] =  df['home_score'] + df ['away_score']
#Sukurti naują stulpelį, kuriame būtų paskaičiuotas įvarčių skirtumas
df['score_diff'] =  df['home_score'] - df ['away_score']
df
```

```python
#Kiek rungtynių baigėsi lygiosiomis?

#df.groupby('goals_difference').count()
df[(df['score_diff'] == 0)].groupby ('score_diff') [['date']].count()
```

```python
#Kokiai šaliai pralošėme didžiausiu ivarčių skirtumu, žaisdami namuose, kada tai įvyko, kokiame mieste vyko rungtynės?

df[ (df['home_team'] == 'Lithuania') & (df['score_diff'] == min(df['score_diff'] ) ) ]
```

```python
#Kiek rungtynių Lietuva pralaimejo žaisdama Lietuvoje?

df[(df['home_team'] == 'Lithuania') & (df['score_diff'] < 0)].count()
```

```python
#Kiek rungtynių Lietuviai sužaidė lygiosiomis namuose?

df[(df['home_team'] == 'Lithuania') & (df['score_diff'] == 0)].count()
```

```python
#Kiek rungtynių Lietuviai sužaidė lygiosiomis išvykoje?
df[(df['away_team'] == 'Lithuania') & (df['score_diff'] == 0)].count()
```

```python
#Kiek rungtynių Lietuva pralaimėjo žaisdama svetur?

df[(df['away_team'] == 'Lithuania') & (df['score_diff'] > 0)].count()
```

```python
#Kiek rungtynių Lietuva laimėjo žaisdama svetur?

df[(df['away_team'] == 'Lithuania') & (df['score_diff'] < 0)].count()
```

```python
#Kiek rungtynių Lietuva laimėjo žaisdama Lietuvoje?


df[(df['home_team'] == 'Lithuania') & (df['score_diff'] > 0)].count()
```
