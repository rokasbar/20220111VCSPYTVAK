#kada lietuva imuse daugiausiai ivarciu?

```python
#1 zingsis: Lietuva home ARBA away team
#2 zingsnis: praplesti 

df[ ((df['home_team']== 'Lithuania') & (df['home_score'] == max(df['home_score']))) \
     |  ( (df['away_team'] =='Lithuania') & (df['away_score'] == max(df['away_score'])))] 

#Gavom viena lentele kur maximalus ivarciu skaicius isvykoj ir namuose
```

```python
#galim tapati pamodeliuoti su atimtim
#KUR LIETUVA ISVYKOJ LAIMEJO DIDZIAUSIU SKIRTUMU

df['score_diff'] = df['home_score'] - df['away_score']
df
df[ (df['away_team'] == 'Lithuania') & (df['score_diff'] == min(df['score_diff'] ) ) ]
```
